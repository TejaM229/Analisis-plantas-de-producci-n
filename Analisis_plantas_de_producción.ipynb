```python
# Importar la librerías
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# URL del archivo CSV que contiene los datos de producción
url = 'https://raw.githubusercontent.com/TejaM229/Analisis-plantas-de-producci-n/main/ventas-1-65d374724924b571259206.csv'
# Utilizar la función read_csv() de pandas para leer el archivo CSV desde la URL y cargarlo en un DataFrame
produccion_df = pd.read_csv(url)
```

```python
print(produccion_df.head(5)) # Este comando imprime las primeras 5 filas del DataFrame
print()
print(produccion_df.tail(5)) # Este comando imprime las últimas 5 filas del DataFrame
```

```python
# Contar la cantidad de productos de un tipo específico
conteo_tipo = produccion_df['tipo'].value_counts()
print("Conteo por tipo:")
print(conteo_tipo)
print()
# Contar la cantidad de productos de un color específico
conteo_color = produccion_df['color'].value_counts()
print("Conteo por color:")
print(conteo_color)
print()
# Contar la cantidad de productos de una talla específica
conteo_talla = produccion_df['talla'].value_counts()
print("Conteo por talla:")
print(conteo_talla)
print()
# Contar la cantidad de productos de una planta específica
conteo_planta = produccion_df['planta'].value_counts()
print("Conteo por planta:")
print(conteo_planta)
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
# Datos
tipos = conteo_tipo.index
cantidad_tipos = conteo_tipo.values

colores = conteo_color.index
cantidad_colores = conteo_color.values

tallas = conteo_talla.index
cantidad_tallas = conteo_talla.values

plantas = conteo_planta.index
cantidad_plantas = conteo_planta.values
print()
# Gráfico de barras para tipo
plt.figure(figsize=(10, 5))
plt.bar(tipos, cantidad_tipos, color='skyblue')
plt.xlabel('Tipo de producto')
plt.ylabel('Cantidad')
plt.title('Conteo por tipo de producto')
plt.xticks(rotation=45, ha='right')
plt.show()
print()
# Gráfico de barras para color
plt.figure(figsize=(10, 5))
plt.bar(colores, cantidad_colores, color='lightblue')
plt.xlabel('Color de producto')
plt.ylabel('Cantidad')
plt.title('Conteo por color de producto')
plt.xticks(rotation=45, ha='right')
plt.show()
print()
# Gráfico de barras para talla
plt.figure(figsize=(10, 5))
plt.bar(tallas, cantidad_tallas, color='skyblue')
plt.xlabel('Talla de producto')
plt.ylabel('Cantidad')
plt.title('Conteo por talla de producto')
plt.xticks(rotation=45, ha='right')
plt.show()
print()
# Gráfico de barras para planta
plt.figure(figsize=(10, 5))
plt.bar(plantas, cantidad_plantas, color='lightblue')
plt.xlabel('Planta de producción')
plt.ylabel('Cantidad')
plt.title('Conteo por planta de producción')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Definir el diccionario de ventas
ventas = {
    'planta': ['TelaTech Factory', 'FiberWorks Manufacturing', 'SilkStream Textiles', 'CottonCraft Works', 'LinenLoom Productions', 'ThreadTide Textile Mill', 'WeaveWave Fabrics', 'YarnYield Manufacturing', 'DyeDreams Factory', 'KnitNest Textile Co.'],
    'precio': [4.55, 4.34, 4.27, 4.13, 4.9, 4.34, 4.72, 4.3, 4.51, 4.67]
}
# Crear el DataFrame ventas_df
ventas_df = pd.DataFrame(ventas)
# Imprimir el DataFrame ventas_df
print(ventas_df)
```

```python
# Calcula el recuento de ocurrencias de cada planta en produccion_df
conteo_planta = produccion_df['planta'].value_counts()
# Asigna el precio correcto a cada planta en ventas_df
precio_por_planta = ventas_df.set_index('planta')['precio']
# Multiplica los recuentos de ocurrencias por el precio correspondiente
costo_tplantas = conteo_planta * precio_por_planta
# Combina las series en un nuevo DataFrame
resultado_df = pd.DataFrame({'Conteo por planta': conteo_planta, 'Precio por planta': precio_por_planta,'Costo por planta': costo_tplantas})
# Imprime el DataFrame combinado
print("Analisis planta:")
print()
print(resultado_df)
# Calcular máximo, mínimo y rango de precios finales
planta_maximo = resultado_df['Costo por planta'].max()
planta_minimo = resultado_df['Costo por planta'].min()
rango_planta = planta_maximo - planta_minimo
print()
print("Rango costo por planta:", rango_planta)
print()
# Mostrar el precio máximo y su lugar correspondiente
print("Costo máximo:", planta_maximo)
# Mostrar el precio mínimo y su lugar correspondiente
print("Costo mínimo:", planta_minimo )
```

```python
# Obtener los nombres de las plantas
plantas = resultado_df.index

# Obtener los costos por planta
costos_por_planta = resultado_df['Costo por planta']

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(plantas, costos_por_planta, color='skyblue')

# Agregar etiquetas y título
plt.xlabel('Planta')
plt.ylabel('Costo por planta')
plt.title('Costo por planta para cada planta')

# Rotar las etiquetas del eje x para mejor visualización si son demasiadas
plt.xticks(rotation=45, ha='right')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# Definir el diccionario costo_tplantas
costo_tplantas = {
    "CottonCraft Works": [189.98],
    "DyeDreams Factory": [234.52],
    "FiberWorks Manufacturing": [256.06],
    "KnitNest Textile Co.": [205.48],
    "LinenLoom Productions": [294],
    "SilkStream Textiles": [196.42],
    "TelaTech Factory": [227.5],
    "ThreadTide Textile Mill": [203.98],
    "WeaveWave Fabrics": [212.4],
    "YarnYield Manufacturing": [215]
}
# Convierte el diccionario "costo_tplantas" en un DataFrame de Pandas llamado "costo_tplantas_df"
costo_tplantas_df = pd.DataFrame(costo_tplantas)
# Imprime el DataFrame "costo_tplantas_df"
print("Costo plantas:")
print()
print(costo_tplantas_df)
# Calcular la suma de los resultados "costo_tplantas_df"
suma_costo_tplantas = costo_tplantas_df.sum().sum()
print()
print("Suma de los costos de las plantas:", suma_costo_tplantas)
```

```python
# Define un diccionario llamado "lugares" que contiene tres listas de datos: 'lugares', 'descuento' y 'envio'.
lugares = {
    "lugares": ["EcoChic Boutique", "Vintage Vogue Emporium", "FitFrenzy Activewear", "Little Sprouts Fashion House", "Local Threads Collective"],
    "descuento": [30, 35, 33, 30, 40],
    "envio": [3, 4, 3, 3, 5]
}
# Convierte el diccionario "lugares" en un DataFrame de Pandas llamado "lugares_df".
lugares_df = pd.DataFrame(lugares)
# Imprime el DataFrame "lugares_df".
print(lugares_df)
```

```python
# Restar el descuento y sumar el costo de envío correspondiente para cada planta
for index, row in lugares_df.iterrows():
    descuento = row['descuento'] / 100
    envio = row['envio'] / 100
    lugares_df.loc[index, 'precio final'] = suma_costo_tplantas - (suma_costo_tplantas * descuento) + (suma_costo_tplantas * envio)

print("Precios finales lugares de venta:")
print()
print(lugares_df)
# Calcular máximo, mínimo y rango de precios finales
precio_maximo = lugares_df['precio final'].max()
precio_minimo = lugares_df['precio final'].min()
rango_precios = precio_maximo - precio_minimo
print()
print("Rango de precios:", rango_precios)
print()
# Mostrar el precio máximo y su lugar correspondiente
lugar_precio_maximo = lugares_df.loc[lugares_df['precio final'].idxmax(), 'lugares']
print("Precio máximo:", precio_maximo, "en", lugar_precio_maximo)
print()
# Mostrar el precio mínimo y su lugar correspondiente
lugar_precio_minimo = lugares_df.loc[lugares_df['precio final'].idxmin(), 'lugares']
print("Precio mínimo:", precio_minimo, "en", lugar_precio_minimo)
```

```python
# Gráfico de barras de los precios finales por lugar de venta
plt.figure(figsize=(10, 6))
plt.bar(lugares_df['lugares'], lugares_df['precio final'], color='skyblue')
plt.title('Precios Finales por Lugar de Venta')
plt.xlabel('Lugar de Venta')
plt.ylabel('Precio Final')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Anotaciones para el precio máximo y mínimo
plt.text(lugar_precio_maximo, precio_maximo, f'Máximo: {precio_maximo}', ha='center', va='bottom')
plt.text(lugar_precio_minimo, precio_minimo, f'Mínimo: {precio_minimo}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
